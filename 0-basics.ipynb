### Stack and Concat

Stack and Concat are two ways to combine tensors.
- Stack: creates a new dimension
- Concat: combines tensors along an existing dimension

````python
x = tf.constant([[1, 2], [3, 4]], dtype=tf.int32)
y = tf.constant([[5, 6], [7, 8]], dtype=tf.int32)

print("x:", x, end="\n\n")
print("y:", y, end="\n\n")

print("CONCAT 0:", tf.concat([x, y], axis=0), end="\n\n")
print("CONCAT 1:", tf.concat([x, y], axis=1), end="\n\n")
print("STACK 0:", tf.stack([x, y], axis=0), end="\n\n")
print("STACK 1:", tf.stack([x, y], axis=1), end="\n\n")
````

# Automatic differentiation  $\nabla f(\mathbf{x})$

TensorFlow implements automatic differentiation (autodiff), which uses calculus to compute gradients.

See the [Automatic differentiation guide](https://www.tensorflow.org/guide/autodiff) for more information.

In this example $z = x^2 + y^2$ and we want to compute the gradient of $z$ with respect to $x$ and $y$.

So, then $ \nabla z = \left( \frac{\partial z}{\partial x}, \frac{\partial z}{\partial y} \right) = \left( 2x, 2y \right)$

````python
x = tf.Variable(1.0)
y = tf.Variable(-1.0)

def f(x, y):
  return x**2 + y**2

with tf.GradientTape() as tape:
  z = f(x, y)

g_f = tape.gradient(z, [x, y])

print(g_f, end="\n\n")
print("grad(z) for (x,y)=(1,-1): ", [t.numpy() for t in g_f])
````

````python
w = tf.Variable(tf.random.normal((3, 2)), name='w')
b = tf.Variable(tf.zeros(2, dtype=tf.float32), name='b')
x = [[1., 2., 3.]]

with tf.GradientTape(persistent=True) as tape:
  y = x @ w + b
  loss = tf.reduce_mean(y**2)

[dl_dw, dl_db] = tape.gradient(loss, [w, b])

print(w.shape)
print(dl_dw.shape)
````

Here is the gradient calculation again, this time passing a dictionary of variables:

````python
my_vars = {
    'w': w,
    'b': b
}

grad = tape.gradient(loss, my_vars)
grad['b']
````

## Eager Execution, Graphs and `tf.function`

TensorFlow 2.0 introduced *"Eager Execution"*, which means that TensorFlow operations are executed immediately, rather than being added to a graph and executed later. This makes TensorFlow easier to use and debug.

BUT if you want to use graphs for performance reasons, you can use [**tf.function**](https://www.tensorflow.org/api_docs/python/tf/function) to convert **a Python function** to a TensorFlow graph function that can be executed more efficiently and that can be exported to a SavedModel.

[https://www.tensorflow.org/guide/basics#graphs_and_tffunction](https://www.tensorflow.org/guide/basics#graphs_and_tffunction)

````python
@tf.function
def my_func(x):
  print('Tracing.\n')
  return tf.reduce_sum(x)


x = tf.constant([1, 2, 3])
print(my_func(x))

x = tf.constant([10, 9, 8])
print(my_func(x))

x = tf.constant([10.0, 9.1, 8.2], dtype=tf.float32)
print(my_func(x))
````

On subsequent calls TensorFlow only executes the optimized graph, skipping any non-TensorFlow steps. Below, note that my_func doesn't print tracing since print is a Python function, not a TensorFlow function.

## Modules, layers, and models  [(ref)](https://www.tensorflow.org/guide/basics#modules_layers_and_models)

`tf.Module` is a class for managing your `tf.Variable` objects, and the `tf.function` objects that operate on them. The `tf.Module` class is necessary to support two significant features:

- You can save and restore the values of your variables using `tf.train.Checkpoint`. This is useful during training as it is quick to save and restore a model's state.
- You can import and export the `tf.Variable` values and the `tf.function` graphs using `tf.saved_model`. This allows you to run your model independently of the Python program that created it.

````python
class MyModule(tf.Module):
  def __init__(self, value):
    self.weight = tf.Variable(value)

  @tf.function
  def multiply(self, x):
    return x * self.weight
  

mod = MyModule(3)
out = mod.multiply(tf.constant([1, 2, 3]))
print(out)
````

## Training loops
[https://www.tensorflow.org/guide/basics#training_loops](https://www.tensorflow.org/guide/basics#training_loops)

We'll generate some random data and train a simple model on it.



For more info see [basic training loops](https://www.tensorflow.org/guide/basic_training_loops)

````python
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.figsize'] = [9, 6]

N = 201
x = tf.linspace(-2, 2, N)
x = tf.cast(x, tf.float32)

def f(x):
  y = x**2 + 2*x - 5
  return y

y = f(x) + tf.random.normal(shape=[N])

def plot_preds(x, y, f, title, model = None):
  plt.figure()
  plt.plot(x, y, '.', label='Data')
  plt.plot(x, f(x), label='Ground truth')
  if model is not None:
    plt.plot(x, model(x), label='Predictions')
  plt.title(title)
  plt.legend()

plot_preds(x, y, f, 'Random generated data')
````

````python
class Model(tf.Module):

  def __init__(self):
    # Randomly generate weight and bias terms
    rand_init = tf.random.uniform(shape=[3], minval=0., maxval=5., seed=22)
    # Initialize model parameters
    self.w_q = tf.Variable(rand_init[0])
    self.w_l = tf.Variable(rand_init[1])
    self.b = tf.Variable(rand_init[2])

  @tf.function
  def __call__(self, x):
    # Quadratic Model : quadratic_weight * x^2 + linear_weight * x + bias
    return self.w_q * (x**2) + self.w_l * x + self.b
````

````python
quad_model = Model()

plot_preds(x, y, f, model=quad_model, title='Before training')
````

````python
def mse_loss(y, y_pred):
  return tf.reduce_mean(tf.square(y - y_pred))
````

````python
batch_size = 16
dataset = tf.data.Dataset.from_tensor_slices((x, y))
dataset = dataset.shuffle(buffer_size=x.shape[0]).batch(batch_size)
````

````python
# Set training parameters
epochs = 100
learning_rate = 0.01
losses = []

# Format training loop
for epoch in range(epochs):
  for x_batch, y_batch in dataset:
    with tf.GradientTape() as tape:
      batch_loss = mse_loss(quad_model(x_batch), y_batch)
    # Update parameters with respect to the gradient calculations
    grads = tape.gradient(batch_loss, quad_model.variables)
    for g,v in zip(grads, quad_model.variables):
        v.assign_sub(learning_rate*g)
  # Keep track of model loss per epoch
  loss = mse_loss(quad_model(x), y)
  losses.append(loss)
  if epoch % 10 == 0:
    print(f'Mean squared error for step {epoch}: {loss.numpy():0.3f}')

# Plot model results
print("\n")
plt.plot(range(epochs), losses)
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error (MSE)")
plt.title('MSE loss vs training iterations')
````

````python
plot_preds(x, y, f, model=quad_model, title='After training')
````

## Try the same code with Keras

Keras is a high-level API for TensorFlow that makes it easy to build and train deep learning models.

````python
new_model = tf.keras.Sequential([
    tf.keras.layers.Lambda(lambda xin: tf.stack([xin, xin**2], axis=1)),
    tf.keras.layers.Dense(units=1, kernel_initializer=tf.random.normal)
])

new_model.compile(
    loss=tf.keras.losses.MSE,
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01))

history = new_model.fit(x, y,
                        epochs=100,
                        batch_size=32,
                        verbose=True)

# new_model.save('./my_new_model')

plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylim([0, max(plt.ylim())])
plt.ylabel('Loss [Mean Squared Error]')
plt.title('Keras training progress')
````

````python
plot_preds(x, y, f, model=new_model, title='After Training: Keras')
````
